```python
def convierteMomio(americano):
    if americano < 100:
        return (100/(-americano))+1
    else:
        return (americano+100)/100
    
def encuentra(a,b):
    p1 = 1/(convierteMomio(a))
    p2 = 1/(convierteMomio(b))

    if (p1+p2) < 1:

        # Ganancias
        r1 = convierteMomio(a)*(calculaMontos2(p1,p2)[0]) - calculaMontos2(p1,p2)[0] - calculaMontos2(p1,p2)[1]
        r2 = convierteMomio(b)*(calculaMontos2(p1,p2)[1]) - calculaMontos2(p1,p2)[1] - calculaMontos2(p1,p2)[0]
        
        r1 = round(r1,4)
        r2 = round(r2,4)
          
        return  "Oportunidad de arbitraje.\n" \
                "-------------------------\n" \
                "Casa 1: " + str(calculaMontos2(p1, p2)[0]) + '(' + str(a) + ')' + "\n" \
                "Casa 2: " + str(calculaMontos2(p1, p2)[1]) + '(' + str(b) + ')' + "\n" \
                "Rendimiento entre: " + str(r1 * 100) + "%" + " y " + str(r2 * 100) + "%"
          
    else:
        return 'No hay oportunidad de arbitraje.'

def calculaMontos(p1,p2):
    margen = 1 - (p1 + p2)
    factor = 1/margen
      
    # Porcentajes de asignación
    a1 = p1 * factor
    a2 = p2 * factor
      
    m1 = (a1/(a1 + a2))
    m2 = (a2/(a1 + a2))
      
    return round(m1,4), round(m2,4)
```
